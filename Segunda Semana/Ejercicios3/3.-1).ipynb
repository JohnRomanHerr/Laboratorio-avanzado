```python
import matplotlib.pyplot as plt
import numpy as np
```

# Disco

```python
rad = 15 #Radio
N = 1000#Numero de puntos

#Generar base (azul)
N=1000
c_x=np.random.uniform(-(rad+1),rad+1,N)
c_y=np.random.uniform(-(rad+1),rad+1,N)

#Generar disco
theta = np.random.uniform(0.0, 2.0*np.pi, N)
r = rad * np.sqrt(np.random.uniform(0.0, 1.0, N))
x_c = r * np.cos(theta)
y_c = r * np.sin(theta)

plt.plot(c_x,c_y,'.')
plt.plot(x_c,y_c,'.')
```

```python
def distancia(x_0,x_1,y_0,y_1):
    d = np.sqrt((x_1 - x_0)**2 + (y_1-y_0)**2)
    return d
#contador
dis_DD = [] #DD
dis_RR =[] #RR
dis_DR = []     #DR
```

```python
for i in range(N):
    for j in range(i+1,N):
        d_dd = distancia(c_x[i],c_x[j],c_y[i],c_y[j])
        d_rr = distancia(c_xx[i],c_xx[j],c_yy[i],c_yy[j])
        dis_DD.append(d_dd)
        dis_RR.append(d_rr)
        #print (d_p)
        
#Distancia de los datos a los random
for  i in  range(N):
        for  j  in  range(N):
            d_dr = distancia(c_x[i],c_xx[j],c_y[i],c_yy[j])
            dis_DR.append(d_dr)
```

```python
DD,dato_bin,n = plt.hist((dis_DD),bins=20)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  DD:}$')
plt.show()
```

```python
RR,random_bin,m = plt.hist((dis_RR),bins=20)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  RR:}$')
plt.show()
```

```python
DR,DR_bin,j = plt.hist((dis_DR),bins=20)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  DR:}$')
plt.show()
```

```python
plt.plot(dato_bin[1:],(DD/RR) -1, label=('$\epsilon^{(2)}_{Peebles-Hauser}$=$\dfrac{DD(r)}{RR(r)}-1$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}$')
plt.legend()
```

```python
plt.plot(dato_bin[1:],(DD*RR)/(DR)**2,label=('$\epsilon^{(2)}_{Hamilton}$=$\dfrac{DD(r)RR(r)}{[DR(r)]^2}$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}_{H}$')
plt.legend()
```

```python
plt.plot(dato_bin[1:],1+ DD/RR -DR/RR, label=('$\epsilon^{(2)}_{Landy-Szalay}$=$1+\dfrac{DD(r)}{RR(r)}-\dfrac{DR(r)}{RR(r)}$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}_{LS}$')
plt.legend()
```
