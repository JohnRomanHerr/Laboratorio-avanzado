```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Generaamos 5,000 puntos aleatorios

```python
N = 5000#Numero de puntos
L = 10#Longutud de la caja
c_x = np.random.uniform(0.,L,N ) #Cordenada en x primer arreglo
c_y = np.random.uniform(0.,L,N)  #Cordenada en y primer arreglo
plt.plot(c_x,c_y,'.',label='Datos')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

Calculamos la distancia entre los puntos y construimos su histograma

```python
def distancia(x_0,x_1,y_0,y_1):
    d = np.sqrt((x_1 - x_0)**2 + (y_1-y_0)**2)
    return d
#contador
dis_DD = [] #DD
```

```python
for i in range(N):
    for j in range(i+1,N):
        d_dd = distancia(c_x[i],c_x[j],c_y[i],c_y[j])
        dis_DD.append(d_dd)
```

```python
DD,dato_bin,n = plt.hist((dis_DD),bins=40)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  DD\ para\ 5,000\ datos:}$')
plt.show()
```

Creamos 5 conjuntos de datos aleatorios con 1,000 puntos cada uno, para intentar recrear el histograma de 5,000 puntos

```python
p=5 #Numero de particiones, y por lo tanto veces que se tine que hacer el nuevo arreglo
n=1000     #N/p=5,000/5  es la nueva cantidad de puntos a generar
#contador
dat_x= []
dat_y=[]
```

```python
for i in range (p):
    rn_x=np.random.uniform(0.,L,n)
    rn_y=np.random.uniform(0.,L,n)
    dat_x.append(rn_x)
    dat_y.append(rn_y)
    plt.plot(dat_x[i],dat_y[i],'.')
```

Se hace el histograma con cada grupo de datos y finalmente se suma para obtener el histograma equivalente 

```python
#contador
dis_DD5 = [] #DD
```

```python
for k in range(p):
    for i in range(n):
        for j in range(i+1,n):
            d_dd5 = distancia(dat_x[k][i],dat_x[k][j],dat_y[k][i],dat_y[k][j])
            dis_DD5.append(d_dd5)
```

```python
DD5,dato5_bin,n = plt.hist((dis_DD5),bins=40)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  DD\ para\ 5\ conjuntos\ de\ 1,000\ datos:}$')
plt.show()
```

```python
plt.plot(dato_bin[1:],(DD/DD5) -5, label=('$\epsilon^{(2)}_{Peebles-Hauser}$=$\dfrac{DD(r)}{DD5(r)}-1$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}$')
plt.legend()
```

```python
theta = np.random.uniform(0.0, 2.0*np.pi, 1000)
r = rad * np.sqrt(np.random.uniform(0.0, 1.0, 1000))
x_c = 5.0+ r * np.cos(theta)
y_c = 5.0+ r * np.sin(theta)
for i in range (p):
    plt.plot(dat_x[i],dat_y[i],'.')
    

plt.plot(x_c,y_c,'.',color='k')
```

```python
dis_DDD = [] #DD
```

```python
for i in range(1000):
    for j in range(i+1,1000):
        d_ddd = distancia(x_c[i],x_c[j],y_c[i],y_c[j])
        dis_DDD.append(d_ddd)
```

```python
DDD,datoD_bin,n = plt.hist((dis_DDD),bins=40)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  DDD\ para\ 1,000\ datos,\ en\ el\ disco:}$')
plt.show()
```

```python
#contador
disD_DR = [] #DD
```

```python
for k in range(p):
    for  i in  range(1000):
        for  j  in  range(100):
            dD_dr = distancia(x_c[i],dat_x[k][j],y_c[i],dat_y[k][j])
            disD_DR.append(dD_dr)
```

```python
DR_D,datoDRD_bin,n = plt.hist((disD_DR),bins=40)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  DD:}$')
plt.show()
```

```python
plt.plot(dato_bin[1:],(5*DDD/DD5) -1, label=('$\epsilon^{(2)}_{Peebles-Hauser}$=$\dfrac{DD(r)}{RR(r)}-1$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}$')
plt.legend()
```

```python
plt.plot(dato_bin[1:],(1/25)*(DDD*DD5)/(DR_D)**2,label=('$\epsilon^{(2)}_{Hamilton}$=$\dfrac{DD(r)RR(r)}{[DR(r)]^2}$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}_{H}$')
plt.legend()
```

```python
plt.plot(dato_bin[1:],1+ (1)*DDD/DD5 -(5)*DR_D/DD5, label=('$\epsilon^{(2)}_{Landy-Szalay}$=$1+\dfrac{DD(r)}{RR(r)}-\dfrac{DR(r)}{RR(r)}$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}_{LS}$')
plt.legend()
```
