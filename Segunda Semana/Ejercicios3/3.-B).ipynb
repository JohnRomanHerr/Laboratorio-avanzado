```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def distancia(x_0,x_1,y_0,y_1):
    d = np.sqrt((x_1 - x_0)**2 + (y_1-y_0)**2)
    return d
#contador
dis_DD = [] #DD
dis_RR = [] #RR
dis_DR = [] #DR
```

```python
L=10 #TamaNo de la caja
p=5 #Numero de particiones, y por lo tanto veces que se tine que hacer el nuevo arreglo
n=1000     #N/p=5,000/5  es la nueva cantidad de puntos a generar
#contador
rand_x= []
rand_y=[]
```

```python
for i in range (p):
    rn_x=np.random.uniform(0.,L,n)
    rn_y=np.random.uniform(0.,L,n)
    rand_x.append(rn_x)
    rand_y.append(rn_y)
    plt.plot(rand_x[i],rand_y[i],'.')
```

```python
rad = 4
theta = np.random.uniform(0.0, 2.0*np.pi, n)
r = rad * np.sqrt(np.random.uniform(0.0, 1.0, n))
x_c = 5.0+ r * np.cos(theta)
y_c = 5.0+ r * np.sin(theta)
for i in range (p):
    plt.plot(rand_x[i],rand_y[i],'.')
    

plt.plot(x_c,y_c,'.',color='k')
```

```python
plt.plot(rand_x[4],rand_y[4],'.')
plt.plot(x_c,y_c,'.',color='k')
```

```python
for i in range(n):
    for j in range(i+1,n):
        d_dd = distancia(x_c[i],x_c[j],y_c[i],y_c[j])
        dis_DD.append(d_dd)
```

```python
DD,dato_bin,n = plt.hist((dis_DD),bins=40)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  DD\ para\ 1,000\ datos,\ en\ el\ disco:}$')
plt.show()
```

```python
for k in range(4):
    for  i in  range(1000):
        for  j  in  range(1000):
            d_dr = distancia(x_c[i],rand_x[k][j],y_c[i],rand_y[k][j])
            dis_DR.append(d_dr)
```

```python
print(len(dis_DR))
```

```python
DR,randoms_bin,m = plt.hist((dis_DR),bins=40)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  DR\ para\ los\ 5 conjuntos\ de\ 1,000\ datos:}$')
plt.show()
```

```python
for k in range(4):
    for  i in  range(1000):
        for  j  in  range(1000):
            d_rr = distancia(rand_x[k][i],rand_x[k][j],rand_y[k][i],rand_y[k][j])
            dis_RR.append(d_rr)
```

```python
RR,randomrandom_bin,p = plt.hist((dis_RR),bins=40)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  RR\ para\ los\ 5 conjuntos\ de\ 1,000\ datos,/ entre/ellos:}$')
plt.show()
```

Correlaciones entre los histoggramas (promediadno)

```python
Nest= 1
plt.plot(dato_bin[1:],Nest*(5*DD/RR) -1, label=('$\epsilon^{(2)}_{Peebles-Hauser}$=$\dfrac{1}{N_{est}^{2}}\dfrac{DD(r)}{RR(r)}-1$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}$')
plt.legend()
```

```python
plt.plot(dato_bin[1:],Nest*(5*DD/DR) -1, label=('$\epsilon^{(2)}_{Peebles-Davis}$=$\dfrac{1}{N_{est}^{2}}\dfrac{DD(r)}{DR(r)}-1$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}$')
plt.legend()
```

```python
plt.plot(dato_bin[1:],(5*(DD*RR)/(DR)**2), label=('$\epsilon^{(2)}_{Hamilton}$=$\dfrac{DD(r)RR(r)}{[DR(r)]^2}$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}$')
plt.legend()
```

```python
plt.plot(dato_bin[1:],(1+Nest*(5*DD/RR) - 2*Nest*(DR/RR)), label=('$\epsilon^{(2)}_{Landy-Szalay}$=$1+\dfrac{1}{N_{est}^{2}}\dfrac{DD(r)}{RR(r)}-2\dfrac{1}{N_{est}^{2}}\dfrac{DR(r)}{RR(r)}$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}$')
plt.legend()
```

```python
plt.plot(dato_bin[1:],Nest*(5*DD/RR) -1, label=('$\epsilon^{(2)}_{Peebles-Hauser}$'))
plt.plot(dato_bin[1:],Nest*(5*DD/DR) -1, label=('$\epsilon^{(2)}_{Peebles-Davis}$'))
plt.plot(dato_bin[1:],(5*(DD*RR)/(DR)**2), label=('$\epsilon^{(2)}_{Hamilton}$'))
plt.plot(dato_bin[1:],(1+Nest*(5*DD/RR) - 2*Nest*(DR/RR)), label=('$\epsilon^{(2)}_{Landy-Szalay}$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}$')
plt.legend()
```

Correlaciones entre los histogramas (sin promediar)

```python
plt.plot(dato_bin[1:],Nest*(DD/RR) -1, label=('$\epsilon^{(2)}_{Peebles-Hauser}$'))
plt.plot(dato_bin[1:],Nest*(DD/DR) -1, label=('$\epsilon^{(2)}_{Peebles-Davis}$'))
plt.plot(dato_bin[1:],((DD*RR)/(DR)**2), label=('$\epsilon^{(2)}_{Hamilton}$'))
plt.plot(dato_bin[1:],(1+Nest*(DD/RR) - 2*Nest*(DR/RR)), label=('$\epsilon^{(2)}_{Landy-Szalay}$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}$')
plt.legend()
```
