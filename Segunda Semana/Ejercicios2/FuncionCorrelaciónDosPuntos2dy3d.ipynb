```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# 2D

Generar dos grupos de puntos aleatorios, los cuales unos consideraremos datos y otros aleatorios.

```python
N = 1000#Numero de puntos
L = 10#Longutud de la caja
c_x = np.random.uniform(0.,L/2,N ) #Cordenada en x primer arreglo
c_y = np.random.uniform(0.,L/2,N)  #Cordenada en y primer arreglo
c_xx = np.random.uniform(0.,L/2,N ) #Cordenada en x segundo arreglo
c_yy = np.random.uniform(0.,L/2,N)  #Cordenada en y segundo arreglo
plt.plot(c_x,c_y,'.',label='Datos')
plt.plot(c_xx,c_yy,'.',label='Random')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

Funcion para la distancia

```python
def distancia(x_0,x_1,y_0,y_1):
    d = np.sqrt((x_1 - x_0)**2 + (y_1-y_0)**2)
    return d
#contador
dis_DD = [] #DD
dis_RR =[] #RR
dis_DR = []     #DR
```

```python
for i in range(N):
    for j in range(i+1,N):
        d_dd = distancia(c_x[i],c_x[j],c_y[i],c_y[j])
        d_rr = distancia(c_xx[i],c_xx[j],c_yy[i],c_yy[j])
        dis_DD.append(d_dd)
        dis_RR.append(d_rr)
        #print (d_p)
        
#Distancia de los datos a los random
for  i in  range(N):
        for  j  in  range(N):
            d_dr = distancia(c_x[i],c_xx[j],c_y[i],c_yy[j])
            dis_DR.append(d_dr)
```

Histograma

```python
DD,dato_bin,n = plt.hist((dis_DD),bins=20)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  DD:}$')
plt.show()
```

```python
RR,random_bin,m = plt.hist((dis_RR),bins=20)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  RR:}$')
plt.show()
```

```python
DR,DR_bin,j = plt.hist((dis_DR),bins=20)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  DR:}$')
plt.show()
```

Funciones de correlacion

```python
plt.plot(dato_bin[1:],(DD/RR) -1, label=('$\epsilon^{(2)}_{Peebles-Hauser}$=$\dfrac{DD(r)}{RR(r)}-1$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}$')
plt.legend()
```

```python
plt.plot(dato_bin[1:],(DD*RR)/(DR)**2,label=('$\epsilon^{(2)}_{Hamilton}$=$\dfrac{DD(r)RR(r)}{[DR(r)]^2}$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}_{H}$')
plt.legend()
```

```python
plt.plot(dato_bin[1:],1+ DD/RR -DR/RR, label=('$\epsilon^{(2)}_{Landy-Szalay}$=$1+\dfrac{DD(r)}{RR(r)}-\dfrac{DR(r)}{RR(r)}$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}_{LS}$')
plt.legend()
```

```python
plt.plot(dato_bin[1:],(DD/RR) -1, label=('$\epsilon^{(2)}_{Peebles-Hauser}$=$\dfrac{DD(r)}{RR(r)}-1$'))
plt.plot(dato_bin[1:],(DD*RR)/(DR)**2,label=('$\epsilon^{(2)}_{Hamilton}$=$\dfrac{DD(r)RR(r)}{[DR(r)]^2}$'))
plt.plot(dato_bin[1:],1+ DD/RR -DR/RR, label=('$\epsilon^{(2)}_{Landy-Szalay}$=$1+\dfrac{DD(r)}{RR(r)}-\dfrac{DR(r)}{RR(r)}$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}$')
plt.legend()
```

# 3D

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
N3 = 500#Numero de puntos
L3 = 10#Longutud de la caja
c3_x = np.random.uniform(0.,L3,N3 ) #Cordenada en x
c3_y = np.random.uniform(0.,L3,N3)  #Cordenada en y
c3_z = np.random.uniform(0.,L3,N3)  #Cordenada en z
c3_xx = np.random.uniform(0.,L3,N3 ) #Cordenada en x
c3_yy = np.random.uniform(0.,L3,N3 ) #Cordenada en x
c3_zz = np.random.uniform(0.,L3,N3 ) #Cordenada en x
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(c3_x,c3_y,c3_z)
ax.scatter3D(c3_xx,c3_yy,c3_zz)
ax.set_xlabel('X ')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Caja con dos conjuntos de puntos aleatorios');
```

```python
def distancia(x3_0,x3_1,y3_0,y3_1,z3_0,z3_1):
    d = np.sqrt((x3_1 - x3_0)**2 + (y3_1-y3_0)**2+ (z3_1-z3_0)**2)
    return d
#contador
dis3_DD = [] #DD
dis3_RR =[] #RR
dis3_DR = [] #DR
```

```python
for i in range(N3):
    for j in range(i+1,N3):
        d3_dd = distancia(c3_x[i],c3_x[j],c3_y[i],c3_y[j],c3_z[i],c3_z[j])
        d3_rr = distancia(c3_xx[i],c3_xx[j],c3_yy[i],c3_yy[j],c3_zz[i],c3_zz[j])
        dis3_DD.append(d3_dd)
        dis3_RR.append(d3_rr)

#Distancia de los datos a los random
for  i in  range(N3):
        for  j  in  range(N3):
            d3_dr = distancia(c3_x[i],c3_xx[j],c3_y[i],c3_yy[j],c3_z[i],c3_zz[j])
            dis3_DR.append(d3_dr)
```

```python
DD3,dato3_bin,n = plt.hist((dis3_DD),bins=20)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  DD:}$')
plt.show()
```

```python
RR3,random3_bin,m = plt.hist((dis3_RR),bins=20)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  RR:}$')
plt.show()
```

```python
DR3,DR3_bin,j = plt.hist((dis3_DR),bins=20)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  DR:}$')
plt.show()
```

```python
plt.plot(dato3_bin[1:],(DD3/RR3) -1, label=('$\epsilon^{(2)}_{Peebles-Hauser}$=$\dfrac{DD(r)}{RR(r)}-1$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}$')
plt.legend()
```

```python
plt.plot(dato3_bin[1:],(DD3*RR3)/(DR3)**2,label=('$\epsilon^{(2)}_{Hamilton}$=$\dfrac{DD(r)RR(r)}{[DR(r)]^2}$'))
plt.xlabel('Distancia')
plt.ylabel('$\epsilon^{(2)}_{H}$')
plt.legend()
```
