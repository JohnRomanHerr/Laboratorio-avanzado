```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
%matplotlib inline
```

# Para la funcion a transformar

```python
t = np.linspace(0, 2*np.pi, 500) #Arreglo de  1000 n√∫meros espaciados uniformemente entre 0 y 2Pi
f = 5 # Frecuencia 
A = 10 # Amplitud
Funcion = A * np.cos(2*np.pi*f*t) #Funcion a transformar
```

Definimos la Transformada de Fourier

```python
def TansDeFou(x):
    N=len(x) #Longitud de x
    #N = x.size
    n = np.arange(N) #Nos da los numeros enteros entre [0,N) ya que la tranformada va de 0 a (N-1)
    k = n.reshape((N,1)) #Creamos un arreglo entre los valores N y 1
    expo = np.exp(-2j*np.pi*k*n/N) #Exponencial relacionada a la transformada discreta de Fouerier
    result = np.dot(expo,x)   #Transformada discreta de Fourier
    return result
```

Funcion a transformar

```python
plt.plot(t,Funcion)
plt.xlabel('Tiempo (t)')
plt.ylabel('Amplitud [f(t)]')
plt.title("Funcion Coseno")
plt.show()
```

```python
import scipy.fftpack
#Transfromada con nuestro alforitmo
F = TansDeFou(Funcion)
plt.plot(F,'r', linewidth = 5, label = 'Nuestro algoritmo')
#Transformada con librerias de Python
Y = np.fft.fft(Funcion)
plt.plot(Y,'k--', linewidth = 1, label='Libreria scipy.fftpack')
plt.title("Transformada de Fourier para Coseno")
plt.legend(loc="upper center") 
plt.show()
```
