```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

```

Generamos las coordenadas aleatorias en 2D

```python
N = 1000#Numero de puntos
L = 10#Longutud de la caja
c_x = np.random.uniform(0.,L/2,N ) #Cordenada en x
c_y = np.random.uniform(0.,L/2,N)  #Cordenada en y
plt.plot(c_x,c_y,'.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Distancia 2D

```python
def distancia(x_0,x_1,y_0,y_1):
    d = np.sqrt((x_1 - x_0)**2 + (y_1-y_0)**2)
    return d
#contador
dis_puntos = []
```

Algoritmo para obtener las distancias entre todos los puntos

```python
for i in range(N):
    for j in range(i+1,N):
        d_p = distancia(c_x[i],c_x[j],c_y[i],c_y[j])
        dis_puntos.append(d_p)
        #print (d_p)
```

Histograma

```python
c_xy, hist,n = plt.hist((dis_puntos),bins=30)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  2D:}$')
#plt.axis([40, 160, 0, 0.03])
#plt.grid(True)
plt.show()
```

# Distancia a 3D

```python
from mpl_toolkits.mplot3d import Axes3D
```

Generar puntos aleatorios en 3D

```python
N3 = 500#Numero de puntos
L3 = 10#Longutud de la caja
c3_x = np.random.uniform(0.,L3,N3 ) #Cordenada en x
c3_y = np.random.uniform(0.,L3,N3)  #Cordenada en y
c3_z = np.random.uniform(0.,L3,N3)  #Cordenada en z

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(c3_x,c3_y,c3_z)
ax.set_xlabel('X ')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Caja con conjuntos de puntos aleatorios');
```

Distancia en 3D

```python
def distancia(x3_0,x3_1,y3_0,y3_1,z3_0,z3_1):
    d = np.sqrt((x3_1 - x3_0)**2 + (y3_1-y3_0)**2+ (z3_1-z3_0)**2)
    return d
#contador
dis3_puntos = []
```

```python
for i in range(N3):
    for j in range(i+1,N3):
        d_p = distancia(c3_x[i],c3_x[j],c3_y[i],c3_y[j],c3_z[i],c3_z[j])
        dis3_puntos.append(d_p)
        #print (d_p)
```

```python
c3_xy, hist3,m = plt.hist((dis3_puntos),bins=30)
plt.xlabel('Distancia')
plt.ylabel('Numeor de puntos')
plt.title('$\mathrm{Histograma\  3D:}$')
plt.show()
```
